```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_training(history, title='VGG16'):
    acc = history['accuracy']
    val_acc = history['val_accuracy']
    loss = history['loss']
    val_loss = history['val_loss']
    miou = history['mean_io_u']
    val_miou = history['val_mean_io_u']
    
    print(title)
    print('Max accuracy:', max(acc))
    print('Max val accuracy:', max(val_acc))
    print('Min loss:', min(loss))
    print('Min val loss:', min(val_loss))
    print('Max mIoU:', max(miou))
    print('Max val mIoU:', max(val_miou))
    
    epochs = range(len(acc))
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_figheight(5)
    fig.set_figwidth(15)
    
    fig.suptitle(title)

    ax1.plot(epochs, acc, 'r.', label='train_acc')
    ax1.plot(epochs, val_acc, 'r', label='val_acc')
    ax1.set_title('Training and validation accuracy')
    ax1.legend()
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Accuracy')
    ax1.set_ylim([0.6, 1])
    
    ax2.plot(epochs, loss, 'r.', label='train_loss')
    ax2.plot(epochs, val_loss, 'r-', label='val_loss')
    ax2.set_title('Training and validation loss')
    ax2.legend()
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Loss')
    ax2.set_ylim([0, 1.5])
    
    plt.figure(figsize=(7,5))
    ax3 = plt.axes()
    ax3.plot(epochs, miou, 'r.', label='train_mIoU')
    ax3.plot(epochs, val_miou, 'r', label='val_mIoU')
    ax3.set_title('Training and validation mIoU')
    ax3.legend()
    ax3.set_xlabel('Epochs')
    ax3.set_ylabel('mIoU')
    ax3.set_ylim([0.24, 0.27])
    
    return ax1,ax2,ax3
```
