```python

def get_histories(var = 'val_loss', y_label = 'loss', helper_fun = np.min, y_min=0, y_max=1, vline=None):
    hists = []
    best_vals = []
    for k in histories.keys():
        history = histories[k]
        history_ft = histories_ft[k]
        
        full_hist = history[var].copy()
        full_hist.extend(history_ft[var])
        
        #loss
        plt.plot(full_hist, label=k)
        if vline:
            plt.axvline(x=vline, color='black')
        plt.ylim([y_min,y_max])
        hists.append(full_hist)
        best_val = helper_fun(full_hist)
        best_vals.append(best_val)
        print(k, var, best_val)
    print('mean:', np.mean(best_vals), 'sd:', np.std(best_vals))
    plt.legend(title="Fold")
    plt.title(var)
    plt.xlabel('epoch')
    plt.ylabel(y_label)
    plt.show()
    return hists
```

```python
def plot_mean_sd_line(data, label="mean validation loss", y_min=0, y_max=1, vline = vline):
    data = np.array(data)
    mean = np.mean(data, axis=0)
    sd = np.std(data, axis=0)
    
    print(label, mean[-1])

    x = range(len(mean))

    plt.plot(x, mean, label=label)
    if vline:
        plt.axvline(x=vline, color='black')
    plt.ylim([y_min,y_max])
    plt.fill_between(x, mean - sd, mean + sd, alpha = 0.2)
```
